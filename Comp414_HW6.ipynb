```
import numpy as np
import math
import matplotlib.pyplot as plt

def f(x):
  return np.log(np.power(math.e, x)+np.power(np.e, -x))

def df(x):
  return (np.power(np.e, 2*x)-1)/(np.power(np.e, 2*x)+1)

def d2f(x):
  return (4*np.power(np.e, 2*x))/(np.power(np.e, 2*x)+1)**2

xmesh = np.linspace(-1, 3
                    , 100)
plt.ylim([-1, 3])
plt.plot(xmesh, f(xmesh))
```

```
x = 1.1

# carry out the Newton step
for i in range(5):
  xnew = x - df(x) / d2f(x)
  x = xnew
xnew = x

# plot approximate function
plt.plot(xmesh, f(xmesh))
plt.plot(xmesh, f(x) + dfx*(xmesh-x) + d2fx*(xmesh-x)**2/2)
plt.plot(x, f(x), "o", color="red")
plt.plot(xnew, f(xnew), "o", color="green")
plt.ylim([-1, 3])

# update
x = xnew
print(x)
```
